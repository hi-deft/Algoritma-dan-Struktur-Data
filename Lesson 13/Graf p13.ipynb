Sebuah jasa pengiriman barang antar kota memiliki jaringan jalur pengiriman sebagai berikut:

*   Dari Kota A dapat langsung mengirim ke Kota B dan Kota D
*   Dari Kota B dapat langsung mengirim ke Kota C dan Kota E
*   Dari Kota C hanya dapat mengirim ke Kota F
*   Dari Kota D dapat mengirim ke Kota B dan Kota E
*   Dari Kota E dapat mengirim ke Kota F
*   Kota F adalah kota tujuan akhir dan tidak mengirim ke kota lain

Berdasarkan informasi tersebut:
Pertanyaan:

1.   Gambarkan graf terarah berdasarkan informasi rute pengiriman.
2.   Buatlah representasi:
*   a. Adjacency List
*   b. Adjacency Matrix (urutkan simpul: A, B, C, D, E, F)
*   c. Edge List
*   d. DFS Traversal
*   e. BFS Traversal
3.   Implementasikan representasi tersebut dalam Python.

```python
import networkx as nx
import matplotlib.pyplot as plt

# Daftar edge dari graf pengiriman barang
edges = [
    ('A', 'B'), ('A', 'D'),
    ('B', 'C'), ('B', 'E'),
    ('C', 'F'),
    ('D', 'B'), ('D', 'E'),
    ('E', 'F')
]

# Buat graf terarah
G = nx.DiGraph()
G.add_edges_from(edges)

# Posisi node otomatis teratur
pos = nx.spring_layout(G)

# Gambar graf
plt.figure(figsize=(8, 6))
nx.draw(G, pos, with_labels=True, node_color='lightblue', node_size=1500, font_size=12, arrows=True)
nx.draw_networkx_edges(G, pos, arrowstyle='->', arrowsize=20)

plt.title("Graf Terarah - Jaringan Pengiriman Barang")
plt.axis('off')
plt.show()
```

```python
# Representasi adjacency list
adj_list = {
    'A': ['B', 'D'],
    'B': ['C', 'E'],
    'C': ['F'],
    'D': ['B', 'E'],
    'E': ['F'],
    'F': []
}

# Menampilkan adjacency list
print("Adjacency List:")
for node in adj_list:
    print(f"{node} -> {adj_list[node]}")
```

```python
# Representasi Adjacency Matrix
nodes = ['A', 'B', 'C', 'D', 'E', 'F']

# Buat matriks kosong 6x6 (jumlah simpul)
size = len(nodes)
adj_matrix = [[0 for _ in range(size)] for _ in range(size)]

# Daftar sisi (edges)
edges = [
    ('A', 'B'), ('A', 'D'),
    ('B', 'C'), ('B', 'E'),
    ('C', 'F'),
    ('D', 'B'), ('D', 'E'),
    ('E', 'F')
]

# Mengisi matriks berdasarkan edges
for u, v in edges:
    i = nodes.index(u)
    j = nodes.index(v)
    adj_matrix[i][j] = 1

# Menampilkan adjacency matrix
print("\nAdjacency Matrix:")
print("   " + "  ".join(nodes))
for i in range(size):
    row = " ".join(str(cell) for cell in adj_matrix[i])
    print(f"{nodes[i]}  {row}")
```

```python
# Implementasi Edge List
edge_list = [
    ('A', 'B'), ('A', 'D'),
    ('B', 'C'), ('B', 'E'),
    ('C', 'F'),
    ('D', 'B'), ('D', 'E'),
    ('E', 'F')
]

# Menampilkan isi edge list
print("\nEdge List:")
for edge in edge_list:
    print(f"{edge[0]} -> {edge[1]}")
```

```python
#Implementasi DFS Traversal
graph = {
    'A': ['B', 'D'],
    'B': ['C', 'E'],
    'C': ['F'],
    'D': ['B', 'E'],
    'E': ['F'],
    'F': []
}

def dfs(graph, start, visited=None):
    if visited is None:
        visited = set()
        print("DFS Traversal:", end=" ")

    if start not in visited:
        print(start, end=" ")
        visited.add(start)
        for neighbor in graph[start]:
            dfs(graph, neighbor, visited)

    if start == 'A':
        print()  # newline at end
print(dfs(graph, 'A'))
```

```python
#Implementasi BFS Traversal
from collections import deque

graph = {
    'A': ['B', 'D'],
    'B': ['C', 'E'],
    'C': ['F'],
    'D': ['B', 'E'],
    'E': ['F'],
    'F': []
}
def bfs(graph, start):
    visited = set()
    queue = deque([start])

    print("BFS Traversal:", end=" ")
    while queue:
        node = queue.popleft()
        if node not in visited:
            print(node, end=" ")
            visited.add(node)
            for neighbor in graph[node]:
                if neighbor not in visited:
                    queue.append(neighbor)
    print()

print(bfs(graph, 'A'))
```
